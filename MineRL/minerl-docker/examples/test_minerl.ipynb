```python
# 1. Check imports
import minerl
import gym
import numpy as np
import matplotlib.pyplot as plt
print('All imports OK')
```

```python
# 2. Check GPU availability (Windows GPU profile only)
import torch
print(f'PyTorch version: {torch.__version__}')
print(f'CUDA available: {torch.cuda.is_available()}')
if torch.cuda.is_available():
    print(f'GPU: {torch.cuda.get_device_name(0)}')
```

```python
# 3. Create a MineRL environment
# This will launch a headless Minecraft instance (takes ~60 seconds first time)
print('Creating MineRLTreechop-v0 environment...')
env = gym.make('MineRLTreechop-v0')
print('Environment created successfully!')
```

```python
# 4. Run a few random steps and render a frame
obs = env.reset()
print(f'Observation keys: {list(obs.keys())}')
print(f'POV shape: {obs["pov"].shape}')

# Display the first frame
plt.figure(figsize=(6, 4))
plt.imshow(obs['pov'])
plt.title('MineRL First Frame')
plt.axis('off')
plt.show()
```

```python
# 5. Random rollout
frames = []
total_reward = 0

for step in range(50):
    action = env.action_space.sample()
    obs, reward, done, info = env.step(action)
    total_reward += reward
    frames.append(obs['pov'])
    if done:
        break

print(f'Ran {len(frames)} steps, total reward: {total_reward}')
env.close()
print('Environment closed.')
```

```python
# 6. Show a grid of frames from the rollout
n = min(10, len(frames))
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, ax in enumerate(axes.flat):
    if i < n:
        ax.imshow(frames[i * (len(frames) // n)])
        ax.set_title(f'Step {i * (len(frames) // n)}')
    ax.axis('off')
plt.suptitle('MineRL Rollout Frames')
plt.tight_layout()
plt.show()
```
