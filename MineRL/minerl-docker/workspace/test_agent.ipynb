```python
import gym
import minerl
import logging

# 1. Setup logging so you can see what's happening in the background
logging.basicConfig(level=logging.DEBUG)

# 2. Create the environment (Navigate is a lightweight one to start with)
# Note: The first time this runs, it might take 5-10 minutes to compile Minecraft
env = gym.make('MineRLNavigateDense-v0')

# 3. Reset the environment to get the first observation
print("Resetting environment... (This may take a while the first time)")
obs = env.reset()

# 4. Run a simple loop of random actions
done = False
steps = 0
while not done and steps < 100:
    # Sample a random action from the environment's action space
    action = env.action_space.sample()
    
    # Apply the action
    obs, reward, done, info = env.step(action)
    
    # Optional: print progress every 20 steps
    if steps % 20 == 0:
        print(f"Step {steps} completed.")
    
    steps += 1

print("Test successful! The environment is working.")
env.close()
```
